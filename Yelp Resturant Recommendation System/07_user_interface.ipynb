```python

import pickle

ctr_frompickle = pickle.load(open(BasePath+'cdl_70_text_de.pkl', 'rb'))
def output_top(model, uid, top_k = 10):
 user_id = list(model.train_set.user_ids)
 item_id = list(model.train_set.item_ids)
 
 print('Start prediction...') 
 item_map = {v:k for k, v in model.train_set.iid_map.items()}
 
 uidx = user_id.index(uid)
 iidx = model.rank(user_idx = uidx)[0][:top_k]
 
 result = [item_map[item_idx] for item_idx in iidx]
 
 return result
```

```python
def predictByName(userid):

  return output_top(ctr_frompickle,userid,100000)
```

```python
def changeRankToResult(rank, city, category):
  
  restaurants = [];
  print(rank)
  for id in rank:
    
    
    business = dic_business[id]
    if(city!=''):
      if(city not in  business['desc1']):
        continue;
    if(category!=''):
      if(category not in business['desc2']):
        continue;
    
    
    restaurants.append(business)

    if len(restaurants)==10:
      break;

  return restaurants
```

```python
@anvil.server.callable
def predict(userid, city, category):
  if(city==None):
    city = ''
  elif(city=='All'):
    city = ''
  if(category==None):
    category = ''
  elif(category=='All'):
    category = ''


  rank = predictByName(userid);
  print(rank)

  restaurants = changeRankToResult(rank, city, category);
 

  return restaurants;
```
